```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Experiment1
Naverage = 34
Ndays = 21
infected = np.zeros((Naverage, Ndays, 4))

for i in range(Naverage):
    with open('Exp1/infected_numbers_' + str(i), 'rb') as f:
        data = pickle.load(f)
    for j in range(4):
        infected[i,:,j] = data['age' + str(j + 1)]
        

population = np.array([448, 868, 6149, 1866])
infected = infected*population[0]/1000

print(infected[0,0,:])
```

```python
infected_average = np.average(infected, axis = 0)
t = np.linspace(0, Ndays - 1, Ndays)

fig, ax = plt.subplots()
plt.errorbar(t, infected_average[:, 0] * 1000 / population[0], label = "0-4", fmt='.r');
plt.errorbar(t, infected_average[:, 1] * 1000 / population[1], label = "5-14", fmt='.g');
plt.errorbar(t, infected_average[:, 2] * 1000 / population[2], label = "15-64", fmt='.b');
plt.errorbar(t, infected_average[:, 3] * 1000 / population[3], label = "65+", fmt='.m');
plt.xlabel("t, days")
plt.ylabel("Number of cases per 1000 population")
# plt.legend(fontsize = 16)
plt.xticks(np.arange(0, 22, 2))

plt.show()
# fig.savefig('Exp1.png', dpi=300)
```
